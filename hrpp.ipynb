<h3>Cleaning in Area Type, City, Furnishing Status, Floors, Tenant Preferred and Point of Contact</h3>

<h3>First: Area Type (Converting Super Area to Carpet Area)</h3>
<h5>Approximation by 70% of super area as carpet area</h5>

```python
df.drop([3079,3966],inplace=True)
```

```python
df.head()
```

```python
df = df.rename(columns={'Area Type':'atype','Furnishing Status':'fs','Tenant Preferred':'tenant','Point of Contact':'poc'})
df
```

```python
def calc_size(row):
    if row['atype'] == 'Super Area':
        return 0.7*row['Size']
    else:
        return row['Size']
```

```python
df['Usable_area'] = df.apply(calc_size,axis=1)
df
```

```python
df.drop(['Size','atype'],axis=1,inplace=True)
df
```

```python
df = df.rename(columns={'Usable_area':'size'})
df
```

<h3>Third: Furnishing status (Ordinal Encoding)</h3>

```python
df['fs'].value_counts()
```

```python
def encode_fs(text):
    if text == 'Furnished':
        return 2
    elif text == 'Semi-Furnished':
        return 1
    else:
        return 0
```

```python
df['Furnish_status'] = df['fs'].apply(encode_fs)
df
```

```python
df.drop(['fs'],axis=1,inplace=True)
df.sample(7)
```

<h5>Data Cleaning complete</h5>

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
x = df.drop('Rent',axis=1)
y = df['Rent']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
s1 = ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(sparse=False,drop='first'),[2,3,5])],remainder='passthrough')
```

```python
lr = LinearRegression()

pipe = Pipeline([
    ('s1',s1),
    ('s2',lr)
])

pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score',r2_score(y_test,y_pred))
```

<h3>Decision Tree Regressor</h3>

```python
dtr = DecisionTreeRegressor(max_depth=6,random_state=4,min_samples_split=8)
pipe = Pipeline([
    ('s1',s1),
    ('s2',dtr)
])

pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score',r2_score(y_test,y_pred))
```

<h3>SVR</h3>

```python
svr = SVR(kernel='rbf',C=10,epsilon=0.01)
pipe = Pipeline([
    ('s1',s1),
    ('s2',svr)
])

pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score',r2_score(y_test,y_pred))
```

<h3>Random Forest Regressor</h3>

```python
rfr = RandomForestRegressor(n_estimators=800,random_state=4,max_depth=9,max_samples=0.5,max_features=0.5)
pipe = Pipeline([
    ('s1',s1),
    ('s2',rfr)
])

pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score',r2_score(y_test,y_pred))
```

<h3>Extra Trees Regressor</h3>

```python
etr = ExtraTreesRegressor(n_estimators=800,random_state=3,max_depth=15,max_samples=0.5,max_features=0.5,bootstrap=True)
pipe = Pipeline([
    ('s1',s1),
    ('s2',etr)
])

pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score',r2_score(y_test,y_pred))
```

<h3>XGB Regressor</h3>

```python
xgb = XGBRegressor(n_estimators=100,max_depth=2,learning_rate=0.4)
pipe = Pipeline([
    ('s1',s1),
    ('s2',xgb)
])

pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 Score',r2_score(y_test,y_pred))
```

<h2>Random Forest Regressor gave the best output with accuracy of 83.29%</h2>
